# Comentarios en Python  

A continuación, aprenderás sobre los comentarios en Python y sus tipos con la ayuda de ejemplos.  

Los comentarios son pistas que podemos añadir a nuestro programa para hacer que nuestro código sea más fácil de entender. Veamos un ejemplo.

Supongamos que tenemos un programa para imprimir un texto ingresado por el usuario.

```python
name = input("Ingrese su nombre: ")
print(name)
```

Además, los comentarios son completamente ignorados por el compilador. Por ejemplo:

```python
# Imprime un saludo
print('Hola')
```

Como puedes ver, el comentario # Imprime un saludo, es ignorado por el interprete de Python y solo se ejecuta print('Hola').

Ahora, veamos los diferentes tipos de comentarios en Python.

# Comentario de una sola línea en Python

En Python, usamos el símbolo numeral (#) para escribir un comentario de una sola línea. Por ejemplo.

```python
# declarar una variable
name = 'Juan'
# imprimir la variable name
print(name) # Juan
```

En el ejemplo anterior, hemos utilizado tres comentarios de una sola línea:

* \# declarar una variable
* \# imprimir la variable name  
* \# Juan

El compilador de Python ignora todo lo que aparece después del símbolo #.

Como puedes ver, también podemos escribir comentarios de una sola línea en la misma línea que el código. Por ejemplo:


```python
print(name) # Juan
```

En tales casos, el compilador ejecuta la instrucción de código antes del # e ignora el texto que aparece después del #.
