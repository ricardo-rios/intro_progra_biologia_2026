# Operadores en Python

En este tutorial, aprenderás sobre los operadores en Python con la ayuda de ejemplos.

Los operadores son símbolos que se utilizan para realizar operaciones sobre valores y variables. Por ejemplo:

```python
print (6 + 5)
```

Aquí, `+` es un operador que suma 6 y 5.

# Tipos de Operadores en Python

Los diferentes tipos de operadores utilizados en Python son:

- Operadores Aritméticos  
- Operadores de Comparación  
- Operadores Lógicos  
- Operadores Especiales  

Aprendamos sobre ellos en detalle.

# Operadores Aritméticos

Los operadores aritméticos realizan operaciones matemáticas como suma, multiplicación, resta, etc.

## Lista de Operadores Aritméticos

- **Suma:** `+`  
- **Resta:** `-`  
- **Multiplicación:** `*`  
- **División:** `/`  
- **División Entera:** `//`  
- **Módulo (Residuo):** `%`  
- **Exponente:** `**`


## Ejemplo: Operadores Aritméticos en Python

```python
x = 7
```

```python
# suma
result = x + 5
print('Resulta de la suma:', result)
```

```python
# resta
result = x - 5
print('Resultado de la resta:', result)
```

```python
# multiplicación
result = x * 10
print('Resulta de la multiplicación:', result)
```

```python
# division
result = x / 10
print('Resultado de la division:', result)
```

```python
# Division entera
quotient = x // 2
print('Cociente:', quotient)
```

```python
# Módulo o Residuo de la división entera
remainder = x % 2
print('Residuo de la división entera:', remainder)
```

```python
# Exponente
result = x ** 2
print('Exponente:', result)
```

# Operadores de Comparación

Los operadores de comparación comparan valores entre operandos y devuelven `True` o `False`. Por ejemplo: `>`, `<`, `==`, etc.

A continuación se muestra una tabla con los operadores de comparación.

| Operador | Ejemplo | Significado                              |
|----------|----------|------------------------------------------|
| >        | x > y    | x es mayor que y                        |
| <        | x < y    | x es menor que y                        |
| ==       | x == y   | x es igual a y                          |
| !=       | x != y   | x es diferente de y                     |
| >=       | x >= y   | x es mayor o igual que y                |
| <=       | x <= y   | x es menor o igual que y                |

## Ejemplo: Operadores de Comparación en Python

Los operadores de comparación se utilizan para comparar valores y devuelven `True` o `False` dependiendo del resultado de la comparación.

A continuación, se muestra un ejemplo en Python:


```python
x = 5
y = 6

# Mayor que
print('x is mayor que y:', x > y)
```

```python
# Mayor o igual que
print('x is mayor o igual que y:', x >= y)
```

```python
# Menor que
print('x es menor que y:', x  < y)
```

```python
# Menor o igual que
print('x es menor o igual que y:', x  <= y)
```

```python
# Igual que
print('x es igual que y:', x == y)
```

# Operadores Lógicos

Los operadores lógicos realizan operaciones lógicas y devuelven `True` o `False`.

Algunos de los operadores lógicos son:

- **and** – devuelve `True` si ambos operandos son verdaderos.  
- **or** – devuelve `True` si al menos uno de los operandos es verdadero.  
- **not** – invierte el valor lógico del operando.  

## Ejemplo: Operadores Lógicos en Python


```python
x = False
y = True

# Operación and
result = x and y
print('x and y es', x and y)
```

```python
# Operación or
result = x or y
print('x or y es',x or y)
```

```python
# Operación not
result = not x
print('not x es',not x)
```

En este ejemplo:

- `and` devuelve `True` solo si ambos valores son `True`.
- `or` devuelve `True` si al menos uno de los valores es `True`.
- `not` cambia el valor de `False` a `True`, o de `True` a `False`.


# Operadores Especiales

En Python, existen tipos especiales de operadores como:

- Operador de identidad  
- Operador de pertenencia  

# Operadores de Identidad

Los operadores de identidad se utilizan para verificar si dos valores están ubicados en la misma posición de memoria.  

Los dos operadores de identidad son:

- **is** – devuelve `True` si ambas variables hacen referencia al mismo objeto.  
- **is not** – devuelve `True` si ambas variables no hacen referencia al mismo objeto.  

## Ejemplo: Operadores de Identidad en Python


```python
# enteros
x1 = 7
y1 = 7

# cadenas
x2 = 'Python'
y2 = 'Python'
```

```python
print(x1 is y1) # True
```

```python
print(x2 is not y2) # False
```

Aquí:

- `x1` y `y1` son enteros iguales e idénticos. Por lo tanto, `x1 is y1` devuelve `True`.  
- `x2` y `y2` son cadenas iguales e idénticas. Por lo tanto, `x2 is not y2` devuelve `False`.  


# Operadores de Pertenencia

Los operadores de pertenencia verifican si un valor determinado se encuentra dentro de una secuencia (cadena).

Los operadores de pertenencia son:

- **in** – devuelve `True` si el valor está en la secuencia.  
- **not in** – devuelve `True` si el valor no está en la secuencia.  

## Ejemplo: Operadores de Pertenencia en Python


```python
# Operador de pertenencia con cadenas
x = 'Python'
print('P' in x) # True
print('P' not in x) # False
```
