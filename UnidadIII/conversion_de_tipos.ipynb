# Conversión de Tipos en Python

En este tutorial aprenderás sobre la conversión de tipos (conversión implícita y explícita) con la ayuda de ejemplos.

Python tiene soporte para diferentes tipos de datos como `int`, `float`, `str`, etc. La conversión de tipos nos permite transformar un tipo de dato en otro.

Existen dos tipos de conversión de datos en Python:

- Conversión implícita de tipos  
- Conversión explícita de tipos  


## Conversión implícita de tipos

En la conversión implícita, Python convierte automáticamente un tipo de dato en otro sin la intervención del usuario.

### Ejemplo: Suma de dos números


```python
# int data
cost = 12000
# float data
vat = 1600.45
total = cost + vat
print(total)
```

Aquí, los tipos de datos de las variables cost y vat son diferentes. Estamos intentando sumar estos dos tipos de datos.

En este caso, Python convierte automáticamente 12000 (int) a 12000.0 (float) antes de realizar la suma.


### Ejemplo: Suma de String e Integer

```python
num_int = 123
num_str = "456"
print(num_int+num_str)
```

Como puedes observar, el programa produce un error. Este es un ejemplo clásico donde la conversión implícita falla en Python.

Python no puede convertir automáticamente estos dos tipos de datos al mismo tipo. Para solucionar este problema, utilizamos la conversión explícita.

## Conversión explícita de tipos

En la conversión explícita, convertimos manualmente un tipo de dato en otro. Esta conversión también se conoce como typecasting, ya que cambia el tipo de dato de los objetos.

Existen funciones predefinidas como int(), float(), str(), etc., que permiten realizar la conversión explícita.

### Ejemplo: Convertir String a Integer

```python
num_int = 123
num_str = "456"
# converting num_str to integer
# converting "456" to 456
num_str = int(num_str)
print(num_int + num_str)
# Output: 579
```

Aquí convertimos el entero 123 a su string equivalente. El resultado es la concatenación de "123" y "456".

### Ejemplo: Error en conversión explícita (Integer a String)

```python
num_int = 123
num_str = "456a"
# trying to covert num_str to integer
# Error! cannot convert "456a" to its equivalent integer
num_str = int(num_str)
print(num_int + num_str)
```

El programa anterior produce un error porque Python no puede convertir "456a" a su entero equivalente (ya que contiene un carácter). En estos casos, Python genera una excepción.
