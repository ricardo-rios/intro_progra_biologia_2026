# Formato de cadenas


Para formatear cadenas en Python a partir de la versión 3.6 se introdujo f-Strings. También llamado formatted string literals ó Formateo literal de cadenas, f-Strings tiene una sintaxis simple y fluida que hará más sencillo el darle formato a cadenas de texto.

En este tutorial, aprenderás sobre f-strings en Python y diferentes formas de darle formato a cadenas de texto.


## ¿Qué son f-Strings en Python?

En Python, una cadena de texto normalmente se escribe entre comillas dobles ("") o comillas simples (''). Para crear f-strings, solo tienes que agregar la letra f o F mayúscula antes de las comillas.


## Como imprimir variables usando f-strings en Python

Si quieres mostrar variables utilizando f-strings, solo tienes especificar el nombre de las variables entre llaves {}. Y al ejecutar tu código, todos los nombres de las variables serán remplazados con sus respectivos valores.

En caso de tener multiples variables en tu cadena de texto, cada variable necesita llaves propias {}.

Aquí puedes ver la sintaxis:


f"Esta es una f-string {nombre_var} y {nombre_var}."


## Ejemplo de una f-string.


Tienes dos variables lenguaje y escuela, entre llaves, dentro de una f-string.



```python
lenguaje = "Python"
escuela = "Escuela de Biología"
print(f"Estoy aprendiendo {lenguaje} en {escuela}.")
```

Puedes ver como las variables  lenguaje y escuela han sido reemplazadas con Python y Escuela de Biología, respectivamente.


## Como evaluar expresiones en Python con f-Strings

Como las f-Strings son evaluadas al momento de ejecución (cuando se está ejecutando el código Python), bien podrías usar una f-string para evaluar expresiones válidas.

En el siguiente ejemplo. num1 y num2 son nuestras variables, para obtener el producto de estas variables únicamente basta con escribir la siguiente expresión num1 * num2 dentro de llaves.

```python
num1 = 83
num2 = 9
print(f"El producto de {num1} y {num2} es {num1 * num2}.")
```

Puedes ver como num1 * num2 es reemplazado  por el producto de num1 y num2 a la hora de ejecutar el código.

En cualquier f-string,  {var_nombre} y {expresion} se sustituirán  por los valores que representan  o por el valor resultante de una operación en el momento de ejecución (cuando se está ejecutando el código Python).

### Introducción al Formateo Numérico en Python

Cuando trabajamos con números decimales, muchas veces necesitamos controlar cuántos decimales se muestran.

Por ejemplo, en contextos financieros no es apropiado mostrar muchos decimales como:

25.33232

En su lugar, queremos algo como:

Para lograr esto, usamos especificadores de formato dentro de f-strings o con la función format().

**Especificador .2f**

El especificador:

.2f

Significa:

* El punto indica que vamos a definir precisión decimal
* El dos indica el número de decimales
* La f indica el formato de número decimal (flotante)

A continuación mostramos un ejemplo:


```python
p = 25.33232
print(f"{p:.2f}")
```

Podemos cambiamos el número de decimales:

```python
print(f"{p:.3f}")
```

### Combinando Decimales con Alineación

Podemos combinar el control de decimales con alineación dentro del mismo formato.

Sintaxis general:

f"{valor:<10.2f}"

Esto significa:

* <  alineación a la izquierda

* 10  ancho total del campo

* .2f mostrar 2 decimales

### Ejemplo práctico: Factura con alineación y decimales


```python
p = 25.33232
x = p * 0.13
t = p + x

print(f"{'Precio:':<15}{'$':<1}{p:.2f}")
print(f"{'IVA:':<15}{'$':<1}{x:.2f}")
print(f"{'Total:':<15}{'$':<1}{t:.2f}")
```

### ¿Por qué es importante?

* Hace que los resultados sean más profesionales
* Mejora la legibilidad
* Es fundamental en reportes financieros y científicos
* Permite controlar precisión y presentación sin modificar el valor original
