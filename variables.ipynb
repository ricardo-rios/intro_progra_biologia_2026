# Variables en Python

A continuación, aprenderás sobre las variables en Python con la ayuda de ejemplos.

En programación, una **variable** es un contenedor que se utiliza para almacenar valores. Por ejemplo:


```python
age = 25
```

Aquí:

- **age** es una variable,  
- **25** es el valor almacenado en la variable **age**, y  
- **=** es el operador utilizado para asignar un valor a una variable.  


Las variables pueden almacenar diferentes tipos de datos. En este ejemplo, la variable **age** está almacenando un valor de tipo entero.


# Imprimir variables en Python

Podemos imprimir una variable fácilmente usando la función **print()**. Por ejemplo,



```python
age = 25
# imprimir la variable age
print(age) # 25
```

Ten en cuenta que si usamos comillas '' en la instrucción **print**, Python lo considera como una cadena de texto. Por ejemplo,

```python
age = 25
# variable rodeada por comillas
# imprimie la cadena age
print('age')
# Salida: age
```

# Declarar variables antes de usarlas

No podemos usar una variable antes de definirla. Por ejemplo:


```python
age = 25
# Usar el nombre de una variable sin definirla produce error
print(name)
print(age)
```

# Cambiar el valor de una variable

En Python, podemos cambiar el valor almacenado en una variable. Por ejemplo:


```python
city = 'New York'
print(city)  # New York
# Cambiar el valor de la variable city
city = 'California'
print(city)  # California
```

Aquí, el valor de la variable **city** cambia de `'New York'` a `'California'`.

**Nota:** No tienes que definir explícitamente el tipo de variable en Python. Cuando asignamos `'New York'` a la variable **city**, Python declara automáticamente **city** como una cadena de texto.


# Asignar una variable a otra

También podemos asignar una variable a otra. Por ejemplo:


```python
city = 'New York'
destination_city = 'California'
# Asignar una variable a otra
city = destination_city
print(city)  # California
```

Aquí, hemos asignado el valor de **destination_city** a **city** usando el operador **=**.

# Reglas para nombrar variables en Python

Estas son las reglas para nombrar variables:

- Un nombre de variable puede contener letras, dígitos y guion bajo.  
- Las variables no pueden tener símbolos especiales como $, @, #, etc.  
- Los nombres de variables no pueden comenzar con un número. Por ejemplo, **1name**.  

Por cierto, siempre debemos intentar usar nombres de variables significativos. Esto hace que tu código sea más fácil de leer y entender.
