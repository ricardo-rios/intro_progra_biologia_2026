# Introducción

A continuación aprenderás sobre los diferentes tipos de datos que podemos usar en Python con la ayuda de ejemplos.

Existen distintos tipos de datos que almacenamos en variables. Estos diferentes tipos se conocen como tipos de datos. Por ejemplo:

```python
x = 5
y = 'Python'
```

Aquí, x es un tipo de dato entero y y es un tipo de dato cadena (string).

Dado que todo en Python es un objeto, los tipos de datos son clases y las variables son instancias (objetos) de esas clases.

Para x = 5, x es una instancia de la clase int.  

## Identificar el tipo de dato

Nota: usamos el método type() para identificar el tipo de dato de una variable. Por ejemplo:



```python
x = 5
print(x)
print(type(x)) # Output: <class 'int'>
```

# Tipos de datos comunes en Python

Algunos de los tipos de datos más usados en Python son:

Números en Python (int, float)

Cadenas en Python (str)


# Int

El tipo de dato entero int representa números enteros positivos o negativos (sin fracciones ni decimales). Por ejemplo:

```python
# número positivo
x = 45
print(x)
print(type(x)) # <class 'int'>
```

```python
# número negativo
x = -3
print(x)
print(type(x)) # <class 'int'>
```

# Float

El tipo de punto flotante float representa números decimales o fraccionarios. Por ejemplo:

```python
x = 5.5
print(x)
print(type(x)) # <class 'float'>
x = 0.002
print(x)
print(type(x)) #<class 'float'>
```

# String

El tipo de dato cadena (str) representa caracteres alfanuméricos. Por ejemplo:


```python
x = 'Car'
print(x)
print(type(x)) # <class 'str'>
x = 'abc12'
print(x)
print(type(x)) # <class 'str'>
x = 'Programando en Python'
print(x)
print(type(x)) # <class 'str'>
```

# Booleanos

El tipo de dato booleano representa uno de dos valores: True o False. Por ejemplo:



```python
x = True
print(type(x)) # <class 'bool'>
```

# None

En Python, None significa ausencia de valor. Por ejemplo:

```python
x = None
print(type(x)) # <class 'NoneType'>
```
